##### reverse() does its job directly on the list
but it returns nothing (returns None)!
So when you print it → you get None!

```python
f = [1, 2, 34, 43, 23423, 32, 34213, 123]

f.reverse()    # Step 1 → reverse the list first
print(f)       # Step 2 → then print the list
```

##### Methods that CHANGE the list → return None
→ reverse(), sort(), append(),
   insert(), extend(), clear()

Always do TWO steps:
Step 1 → call the method
Step 2 → print the list
