#### Comparison Operator 

```python
a = 10
b = 5

print(a == b)   # False
print(a != b)   # True
print(a > b)    # True
print(a < b)    # False
print(a >= 10)  # True
print(b <= 5)   # True
```

### Logical Operator 

```python
age = 20
has_id = True

# and → both conditions must be true
print(age >= 18 and has_id == True)   # True

# or → at least one must be true
print(age >= 18 or has_id == False)   # True

# not → flips the result
print(not has_id)                      # False
```

### String Functions
Python have many build in string functions that we use in our program.

```python
name = "aashish budhathoki"

print(name.upper())       # AASHISH BUDHATHOKI
print(name.lower())       # aashish budhathoki
print(name.title())       # Aashish Budhathoki
print(name.capitalize())  # Aashish budhathoki
```

```python
text = "   hello world   "

print(text.strip())    # "hello world"  → removes both sides
print(text.lstrip())   # "hello world   " → removes left only
print(text.rstrip())   # "   hello world" → removes right only
```

```python
text = "I love Python"

print(text.replace("Python", "coding"))  # I love coding
print(text.find("Python"))               # 7 → position where it starts
print(text.count("o"))                   # 2 → counts how many times
```

```python
print("hello".startswith("he"))   # True
print("hello".endswith("lo"))     # True
print("hello123".isalnum())       # True  → only letters and numbers?
print("hello".isalpha())          # True  → only letters?
print("123".isdigit())            # True  → only numbers?
```

```python
text = "Hello Python"

print(len(text))          # 12 → total characters
print(text.split(" "))    # ['Hello', 'Python'] → splits into list
print(" ".join(["Hello", "Python"]))  # Hello Python → joins list
```
