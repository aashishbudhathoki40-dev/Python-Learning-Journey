```python
import random

user_score = 0
com_score = 0
attempts = 3         # only 3 chances!

com = random.randint(1, 10)

print("🎮 Number Guessing Game!")
print(f"You have {attempts} attempts!")
print("━━━━━━━━━━━━━━━━━━━━━━━")

while attempts > 0:
    print(f"\n📊 Scoreboard → You: {user_score} | Computer: {com_score}")
    print(f"Attempts left: {attempts}")

    user_input = int(input("Enter a number: "))

    if com == user_input:
        user_score += 10
        com_score -= 10
        print(f"✅ Correct! You Win!")
        print(f"\n🏆 Final Score → You: {user_score} | Computer: {com_score}")
        break
    else:
        com_score += 10
        user_score -= 10
        attempts -= 1
        print(f"❌ Wrong! Try Again!")

        if attempts == 0:
            print(f"\n💀 Game Over! The number was {com}")
            print(f"🏆 Final Score → You: {user_score} | Computer: {com_score}")
```
