### Identity Operators
is , is not 

```python
# Creating two separate lists with identical contents
list_one = [10, 20, 30]
list_two = [10, 20, 30]

# Creating an alias (pointing to the same list)
list_three = list_one

# Equality check (values)
print(list_one == list_two)       # Output: True

# Identity check (memory locations)
print(list_one is list_two)       # Output: False
print(list_one is list_three)     # Output: True
print(list_one is not list_two)   # Output: True

# Verifying using id()
print(id(list_one) == id(list_three)) # Output: True
```

### Match Case

```python
def check_status(status_code):
    match status_code:
        case 200:
            return "Success"
        case 400:
            return "Bad Request"
        case 404:
            return "Not Found"
        case _:
            return "Unknown Status"
```

### Single line if condition

```python
score = 85
status = "Pass" if score >= 50 else "Fail"
print(status)  # Output: Pass
```

```python
# Normal way (3 lines)
age = 20
if age >= 18:
    print("Adult")
else:
    print("Minor")

# Single line way (1 line) ✅
print("Adult" if age >= 18 else "Minor")
```
